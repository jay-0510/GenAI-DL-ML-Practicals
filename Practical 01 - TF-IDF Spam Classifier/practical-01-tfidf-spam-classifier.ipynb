```python
""""Importing the Libraries"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
)

import pickle
```

```python
"""Loading the Dataset"""
df = pd.read_csv("spam.csv", encoding="latin-1")
df.head()
```

```python
# Data Cleaning
df = df[['v1','v2']]
df.columns=['label','message']
df.head()
```

# Exploratory Data Analysis

```python
## Check how many spam vs ham messages exist - class imbalance matters for evaluation choices
print(df['label'].value_counts())

## Create a new feature: length of each message (spam messages are often longer/shorter in patterns)
df['length']=df['message'].apply(len)
df['length'].describe()
```

```python
# Visualize class distribution to confirm imbalance visually (bar chart)
df['label'].value_counts().plot(kind='bar',title='Class Distribution')
plt.show()

# Visualize message length distribution to see if length could be a useful signal
plt.hist(df['length'],bins=40)
plt.title("Message Length Distribution")
plt.show()
```

# Label Encoding

```python
# Convert text labels (ham/spam) into numeric form (0/1) since ML models need numbers, not strings
encoder=LabelEncoder()
df['label']=encoder.fit_transform(df['label'])
df.head()
```
